```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = cv2.imread('../images/butterfly.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(20, 20))
plt.subplot(2, 2, 1)
plt.title("Original")
plt.imshow(image)
```

```python
smaller = cv2.pyrDown(image)
larger = cv2.pyrUp(smaller)
plt.subplot(2, 2, 2)
plt.title("Smaller")
plt.imshow(smaller)
```

```python
plt.subplot(2, 2, 3)
plt.title("Larger")
plt.imshow(larger)
```
